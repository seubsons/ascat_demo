```python
import folium

# Define example data for wind speed field
lats = [30.0, 35.0, 40.0]
lons = [-130.0, -120.0, -110.0]
wind_speeds = [[10.0, 15.0, 20.0], [5.0, 10.0, 15.0], [0.0, 5.0, 10.0]]

# Create a map centered on a location of your choice
map = folium.Map(location=[37.7749, -122.4194], zoom_start=3)

# Loop through the wind speed data and add a circle marker for each location
for i in range(len(lats)):
    for j in range(len(lons)):
        lat, lon = lats[i], lons[j]
        wind_speed = wind_speeds[i][j]
        # Choose a color based on the wind speed
        if wind_speed < 5:
            color = "green"
        elif wind_speed < 10:
            color = "yellow"
        else:
            color = "red"
        # Add a circle marker with the wind speed as the popup and the color depending on wind speed
        marker = folium.vector_layers.CircleMarker(location=[lat, lon], radius=5, popup=str(wind_speed)+" m/s", color=color, fill=True, fill_opacity=.5, weight=0)
        # Add the marker to the map
        marker.add_to(map)

# Display the map
map
```

```python
import folium
from folium.plugins import HeatMap

# Define example data for wind speed field
lats = [30.0, 35.0, 40.0]
lons = [-130.0, -120.0, -110.0]
wind_speeds = [[10.0, 15.0, 20.0], [5.0, 10.0, 15.0], [0.0, 5.0, 10.0]]

# Reshape the data into a list of tuples containing latitude, longitude, and wind speed
wind_data = []
for i in range(len(lats)):
    for j in range(len(lons)):
        wind_data.append([lats[i], lons[j], wind_speeds[i][j]])

# Create a map centered on a location of your choice
map = folium.Map(location=[37.7749, -122.4194], zoom_start=3)

# Create a heatmap layer with the wind speed data
heatmap = HeatMap(wind_data, name="Wind Speed", min_opacity=0.2, radius=10, blur=7, max_zoom=1)

# Add the heatmap layer to the map
heatmap.add_to(map)

# Display the map
map
```
